```python
def get_staffing():
    
    ret = pd.read_csv("../clean/csv/fte_bilingual_tesol.csv")
    ret["district_code"] = pd.to_numeric(ret["district_code"],errors="coerce")\
    .fillna("0")\
    .astype(int).\
    astype(str).str.zfill(7)
    
    return ret[ret["district_code"] != "0000000"]

get_staffing()
```

```python
def get_ell():
    
    ret = pd.read_csv("../clean/csv/enrollment-trend-ell.csv")
    
    ret["district_code"] = ret["district_code"].astype(str).str.zfill(7)
    ret["yes"] = pd.to_numeric(ret["yes"],errors="coerce").fillna(0)
    ret["no"] = pd.to_numeric(ret["no"],errors="coerce").fillna(0)
    ret["total"] = pd.to_numeric(ret["total"],errors="coerce").fillna(0)


    return ret

get_ell()
```

```python
# Summarize ell to district-level

def summarize_ell(df):
    ret = df.copy()
    
    ret["yes"] = pd.to_numeric(ret["yes"],errors="coerce")
    ret["no"] = pd.to_numeric(ret["no"],errors="coerce")
    ret = ret.groupby(["district_code","year"]).agg({"yes":sum,"no":sum})
    ret = ret.reset_index()
    return ret
    
ell_dist = summarize_ell(ell)

ell_dist.head(10)
```

```python
def get_row(district_code, year):
    matches = ell_dist[(ell_dist["district_code"] == district_code)
                    &(ell_dist["year"] == year)]
    
    if len(matches) < 1:
        return None
    return matches.iloc[0]["yes"],matches.iloc[0]["no"]

get_row("0010011","2016-17")
```

```python
def combine ():
    
    ret = fte.copy()
    
    
    ret = gs.add_drg(ret)
    
    ret["yes"] = ret.apply(lambda x: get_row(x["district_code"],x["year"])[0], axis=1)
    ret["no"] = ret.apply(lambda x: get_row(x["district_code"],x["year"])[1], axis=1)

    ret["st_ratio"] = ret["yes"] / ret["educators_total"]
    ret["school_code"] = ret["district_code"]
    return ret

combine().sort_values(by="st_ratio",ascending=False)
    
```

```python
def drg_report(df):
    
    ret = df.copy()
    
    ret = ret.groupby(["drg","year"]).agg({"st_ratio": np.nanmean})
    
    return ret.reset_index()

drg_report(combine())
```

```python
def state_report(df):
    ret = df.copy()
    
    ret = ret.groupby("year").agg({"st_ratio":np.mean})
    
    ret = ret.reset_index()
    
    return ret

state_report(combine())
```

```python
def merge_all():
    
    drg_df = drg_report(combine())
    state_df = state_report(combine())
    
    def get_drg_ratio(drg, year):
        
        matches = drg_df[(drg_df["drg"] == drg) & (drg_df["year"] == year)]
        
        if len(matches) != 1:
            return None
        
        return matches.iloc[0]["st_ratio"]
        
    
    def get_state_ratio(year):
        
        matches = state_df[state_df["year"] == year]
        
        if len(matches) != 1:
            return None
        
        return matches.iloc[0]["st_ratio"]

    df = combine()
    
    df["drg_ratio"] = df.apply(lambda x: get_drg_ratio(x["drg"],x["year"]),axis=1)
        
    df["state_ratio"] = df.apply(lambda x: get_state_ratio(x["year"]),axis=1)

    
    return df

merge_all()
```
