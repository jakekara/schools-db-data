```python
import pandas as pd
import getschools as gs
```

```python

def get_df(fname = "LANGUAGES_SPOKEN -school 2014-15 x5311"):
    fext = ".xlsx"
    srcdir="../raw/"
    dstdir="../clean/"

    ret = pd.read_excel(srcdir + fname + fext, skiprows=1)
    
    ret = gs.clean_colnames(ret)
    try:
        ret = gs.add_exact_matches(ret)
    except:
        pass
    return ret

```

```python
df = get_df()
df.columns = "district","school","language","total_enrollment","els","school_code","check_sch_col","check_dist_col"
df.head()
```

```python
df.to_csv("../clean/csv/languages_spoken_schools.csv",index=False)
```

```python
def process_district_file(df):
    
    ret = df.copy()
    ret.columns = "district","language","sutdents","els"
    ret = gs.add_district_matches(ret)
    
    return ret
    
dist_df = process_district_file(get_df(fname="DISTRICT - LANGUAGES SPOKEN - district, state 2014-15 x5297"))

```

```python
dist_df.columns = "district","language","students","els","district_code","check_dist_col"
dist_df["school_code"] = dist_df["district_code"]

dist_df.head()
```
