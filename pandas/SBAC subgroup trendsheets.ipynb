```python
import pandas as pd, re
import getschools as gs
```

```python
def get_df(year, group):
    
    ret = pd.read_csv("../scripts/noequals/sbac-" + year + "-" + group + ".csv",skiprows=3,header=None)
    
    ret = gs.merge_01(ret)
    
    ret = gs.clean_colnames(ret)
    
    ret["district_code"] = ret["district_code"].str.replace("\"","")
    ret["school_code"] = ret["school_code"].str.replace("\"","")

    ret["district_code"] = ret["district_code"].astype(str).str.zfill(7)
    ret["school_code"] = ret["school_code"].astype(str).str.zfill(7)
    
    
    ret["year"] = year

    return ret

get_df("2014-15","ELL")
```

```python
years = map(lambda x: "20" + str(x).zfill(2) + "-" + str(x+1).zfill(2),range(14,16))

def make_trend(subgroup):
    
    sheets = []
    
    for y in years:
        print y
        sheets.append(get_df(y,subgroup))
    
    #return len(sheets)
    ret = pd.concat(sheets)
    
    return ret
```

```python
make_trend("ELL")
```

```python

groups = {
    
    "all":"All+Students",
    "ell":"ELL",
    "gender":"Gender",
#    "frpl2":"Reduced Price Meal Eligibility (2-level)",
#    "frpl2":"Free-Reduced Price Meal Eligibility (3-level)",
    "race":"Race",
    "sped":"Special"
}


for g in groups:
    make_trend(groups[g]).to_csv("../clean/csv/sbac-trend-" + g + ".csv",
                                 index=False)
```
