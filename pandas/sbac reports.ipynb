```python
import pandas as pd
import numpy as np
import getschools as gs
```

```python
def get_df(group="all"):
    
    ret = pd.read_csv("../clean/csv/sbac-trend-"+group+".csv")
    
    ret["district_code"] = ret["district_code"].astype(str).str.zfill(7)
    ret["school_code"] = ret["school_code"].astype(str).str.zfill(7)
    
    
    ret = gs.add_drg(ret)
    
    return ret

get_df("race")
```

```python
def drg_report(df, groups=[]):
    ret = df.copy()
    
    def to_numeric(col):
        ret[col] = pd.to_numeric(ret[col], errors="coerce")
        
    to_numeric("smarter_balancedparticipationrate")
    to_numeric('total_numberof_students')
    to_numeric('total_numbertested')
    to_numeric('total_number_with_scored_tests')
    
    
    for c in ['level_1not_met_level_1_not_met_count', 
              u'level_1not_met_pct',
              u'level_2approaching_level_2_approaching_count',
              u'level_2approaching_pct', 
              u'level_3met_level_3_met_count',
              u'level_3met_pct', 
              u'level_4exceeded_level_4_exceeded_count',
              u'level_4exceeded_pct',
              u'level_3_4met_or_exceeded_level_3_4_met_or_exceeded_count',
              u'level_3_4met_or_exceeded_pct']:
        
        to_numeric(c)
        
    ret = ret.groupby(["drg","subject", "year"] + groups).agg({
            'total_numberof_students':np.sum,
            'total_numbertested':np.sum,
            'smarter_balancedparticipationrate':np.median,
            'total_number_with_scored_tests':np.sum,
            
            'level_1not_met_pct':np.median,
            'level_2approaching_pct':np.median,
            'level_3met_pct':np.median,
            'level_4exceeded_pct':np.median,
            'level_3_4met_or_exceeded_pct':np.median
            
        }).reset_index()
    
    return ret

drg_report(get_df("race"),
           groups=["race"])
```

```python
def state_report(df,groups=[]):
    ret = df.copy()
    
    def to_numeric(col):
        ret[col] = pd.to_numeric(ret[col], errors="coerce")
        
    to_numeric("smarter_balancedparticipationrate")
    to_numeric('total_numberof_students')
    to_numeric('total_numbertested')
    to_numeric('total_number_with_scored_tests')
    
    
    for c in ['level_1not_met_level_1_not_met_count', 
              u'level_1not_met_pct',
              u'level_2approaching_level_2_approaching_count',
              u'level_2approaching_pct', 
              u'level_3met_level_3_met_count',
              u'level_3met_pct', 
              u'level_4exceeded_level_4_exceeded_count',
              u'level_4exceeded_pct',
              u'level_3_4met_or_exceeded_level_3_4_met_or_exceeded_count',
              u'level_3_4met_or_exceeded_pct']:
        
        to_numeric(c)
    
    ret = ret.groupby(["year","subject"] + groups).agg({
            'total_numberof_students':np.sum,
            'total_numbertested':np.sum,
            'smarter_balancedparticipationrate':np.median,
            'total_number_with_scored_tests':np.sum,
            
            'level_1not_met_pct':np.median,
            'level_2approaching_pct':np.median,
            'level_3met_pct':np.median,
            'level_4exceeded_pct':np.median,
            'level_3_4met_or_exceeded_pct':np.median
            
        }).reset_index()

    return ret

state_report(get_df())
```

```python
get_df("race").to_csv("../clean/for_db/sbac_race.csv",index=False)
drg_report(get_df("race"),groups=["race"]).to_csv("../clean/for_db/drg_sbac_race.csv",index=False)

```

```python
get_df("all").to_csv("../clean/for_db/sbac_all.csv",index=False)
drg_report(get_df("all")).to_csv("../clean/for_db/drg_sbac_all.csv",index=False)
```
