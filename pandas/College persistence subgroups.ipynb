```python
import pandas as pd, re
import getschools as gs
```

```python
# files look like collegeEntrancePersistence-all.csv

srcdir = "../scripts/noequals/"
dstdir = "../clean/csv/"
def fname(group):
    return "collegeEntrancePersistence-" + group + ".csv"

def get_df(group):
    global srcdir
    ret = pd.read_csv(srcdir + fname(group), skiprows=7,header=None)

    ret = gs.merge_01(ret)
    
    ret = gs.clean_colnames(ret)
    
    ret["district_code"] = ret["district_code"].str.replace("\"","")
        
    ret["school_code"] = ret["school_code"].str.replace("\"","")
    
    def remove_trail(s):
        if (s[-1] == "_"):
            return s[:-1]
        return s
    
    ret.columns = map(remove_trail, ret.columns)

    ret["group"] = group
    
    ret.to_csv(dstdir + fname(group),index=False)
    return ret

get_df("all").head()

```

```python
groups = ["all","ell","frpl",
          "gender","race","sped"]

for g in groups:
    get_df(g)
```
