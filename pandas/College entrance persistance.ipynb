```python
import pandas as pd
import getschools as gs
```

```python
def get_df():
    
    srcdir="../scripts/noequals/"
    dstsrc="../clean/csv/"
    fname="collegeEntrancePersistence.csv"
    ret = pd.read_csv(srcdir + fname,skiprows=7,header=None)
    
    row0 = ret.iloc[0].fillna(method="ffill")
    row1 = ret.iloc[1].fillna("")
    
    cols = []
    for i in range(len(row0)):
        if row0[i] is None:
            cols.append(str(row1[i]))
        else:
            cols.append(str(row0[i]) + "_" + str(row1[i]))

    cols = map(lambda x: x.replace("nan_",""), cols)

    ret = ret.drop(0)
    ret = ret.drop(1)
    ret.columns = cols
    
    
    def remove_trail(s):
        if (s[-1] == "_"):
            return s[:-1]
        return s
        
    ret = gs.clean_colnames(ret)
    
    ret.columns = map(remove_trail, ret.columns)
    
    ret["district_code"] = ret["district_code"].str.replace("\"","")
    ret["school_code"] = ret["school_code"].str.replace("\"","")

    ret.to_csv(dstsrc + fname,index=False)

    return ret

get_df()
```
