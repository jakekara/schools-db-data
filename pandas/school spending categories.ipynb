```python
import pandas as pd
import getschools as gs
import numpy as np
```

```python
def get_df():
    sheetname = "../scripts/noequals/fiscalResources.csv"
    
    ret = pd.read_csv(sheetname,skiprows=4)
    ret["District"] = ret["District"].fillna(method='ffill')
    ret = gs.clean_colnames(ret)
    ret["school"] = ret["district"]
    ret = gs.add_district_matches(ret)
    ret["district_code"] = ret["dist_code"]
    ret = gs.add_drg(ret)
    return ret

get_df()[get_df()["district"] == "Hartford School District"]
```

```python
df = get_df()
print "bad " + str(len(df[df["dist_code"].isnull()]["district"].unique()))
print "good " + str(len(df[df["dist_code"].notnull()]["district"].unique()))
```

```python

def get_total(dist_name,year=None):
    #ret = get_df()
    ret = df.copy()
    
    try:
        ret = ret[(ret["district"] == dist_name) & (ret["type"] == "Total Expenditures")]
    except:
        pass
        
    if year is not None:
        ret = ret[year]
    
    try:
        return float(ret)
    except:
        pass

get_total("Woodbridge School District",year="2014_15")
```

```python
get_total("Fake School District",year="2014_15")
```

```python
def get_pct(dist_name,amount,year="2014_15"):
    try:
        total = get_total(dist_name,year=year)
    except Exception, e:
        print e
        return
    return amount / total
    
print get_pct("Woodbridge School District",132,year="2014_15")
```

```python
def pct_table(year="2014_15"):
    ret = df[["drg","district_code","district","type",year]]
    ret["year"] = year.replace("_","-")
    
    ret.columns = "drg","district_code","district","type","amount","year"
    ret["pct"] = ret.apply(lambda x: get_pct(x["district"],x["amount"],year),axis=1)
    return ret

pct_table()
```

```python
def one_topic(df, topic="Instructional Staff and Services"):
    
    ret = df.copy()
    
    ret = ret[ret["type"] == topic]
    
    return ret
one_topic(pct_table())[one_topic(pct_table())["district"] == "Hartford School District"]
```

```python
def describe(df, topic="Instructional Staff and Services"):
    
    ret = one_topic(df, topic)
    
    return ret["pct"].describe()
    
describe(pct_table())
```

```python
def drg_summary(df, agg):
    return df.groupby("drg").agg(agg)

drg_summary(one_topic(pct_table()).sort_values(by="pct",ascending=False), {"pct":np.median})
```

```python
def state_sum(topic="Instructional Staff and Services",year="2014_15"):
    
    ret = df[df["type"] == topic]
    ret = np.nansum(ret[year])
    
    return ret

state_sum()
```

```python
def state_pct(topic="Instructional Staff and Services",year="2014_15"):
    return state_sum(topic, year) / state_sum("Total Expenditures")

state_pct()
```

```python
def state_pct_summary(year="2014_15"):
    
    ret = []
    for t in df["type"].unique():
        ret.append({"type":t,
                  "pct":state_pct(topic=t,year=year),
                 "year":year})
        
    return pd.DataFrame(ret)[["year","type","pct"]].sort_values(by="pct",ascending=False)

state_pct_summary()
```

```python
def main_frame(year="2014_15"):
    frames = []
    for t in df["type"].unique():
        f = one_topic(pct_table(year=year),topic=t)  
        f["state_pct"] = state_pct(topic=t,year=year)
        frames.append(f)
    ret = pd.concat(frames)
    ret["school_code"] = ret["district_code"]
    
    return ret
#mf = main_frame()
```

```python
mf = main_frame(year="2013_14")
mf
```

```python
def big_mf():
    ret = []
    for y in ["2010_11","2011_12","2012_13","2013_14","2014_15"]:
        ret.append(main_frame(year=y))
        
    return pd.concat(ret)

bmf = big_mf()
bmf
```
