```python
import pandas as pd
import getschools as gs
```

```python
def get_df(suff="all"):
    fname = "PrimaryDisability-" + suff + ".csv"
    srcdir="../scripts/noequals/"
    dstdir="../clean/csv/"
    
    df = pd.read_csv(srcdir + fname,skiprows=2,header=[0,1])
    
    df.columns=["district","district_code",
                "2011_12_count","2011_12_rate",
                "2012_13_count","2012_13_rate",
                "2013_14_count","2011_14_rate",
                "2014_15_count","2012_15_rate",
                "2015_16_count","2012_16_rate"]
    
    df["group"] = suff
    df["district_code"] = df["district_code"].str.replace("\"","")
    df["school_code"] = df["district_code"]
    df.to_csv(dstdir + fname, index=False)
    return df

get_df()
    
```
