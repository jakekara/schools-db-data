```python
import pandas as pd
import numpy as no
import getschools as gs
```

```python
def get_df():
   
    ret = pd.read_excel("../raw/Copy of 420.xlsx", sheetname="Avg Salary Gen Ed Teacher",
                       skiprows=3, heading=0)
    
    ret = gs.clean_colnames(ret)
    
    ret["district_code"] = ret["district_code"].fillna(0).apply(lambda x: str(int(x)))#.str.zfill(7)
    ret["school_code"] = ret["school_code"].fillna(0).apply(lambda x: str(int(x)))#.str.zfill(7)
     
    ret = gs.add_drg(ret)
    return ret

get_df()
```
