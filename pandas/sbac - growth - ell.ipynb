```python
import pandas as pd
import numpy as np
import getschools as gs
```

```python
def get_df():
    
    ret = pd.read_csv("../scripts/noequals/sbac-growth-ell.csv",skiprows=2)
    ret = gs.clean_colnames(ret)
    
    ret["year"] = "2015-16"
    
    ret["district_code"] =  ret["district_code"].apply(lambda x: x.replace("\"",""))
    ret["school_code"] =  ret["school_code"].apply(lambda x: x.replace("\"",""))

    def to_num(x):
        x = x.replace("%","")
        
        return x
    
    def col_numeric(col):
        ret[col] = ret[col].apply(to_num)
        
        ret[col] = pd.to_numeric(ret[col],errors="coerce")
        
    col_numeric("growth_rate")
    
    ret = gs.add_drg(ret)

    
    return ret

get_df()
```

```python
get_df()[(get_df()["growth_rate"].notnull()) & (get_df()["english_learner_status_el_"] == "ELL")]
```

```python
get_df()[(get_df()["growth_rate"].isnull()) & (get_df()["english_learner_status_el_"] == "Non-ELL")]
```

```python
get_df().to_csv("../clean/for_db/sbac_growth_ell.csv",index=False)
```

```python
def drg_report():
    ret = get_df()
    
    ret = ret.groupby(["year","drg","subject","english_learner_status_el_"]).agg({"growth_rate":np.nanmedian}).reset_index()
    
    return ret

drg_report().to_csv("../clean/for_db/drg_sbac_growth_ell.csv", index=False)
```

```python
def state_df():
    
    ret = pd.read_csv("../raw/sbac-growth-state.csv",skiprows=2)
    
    ret = gs.clean_colnames(ret)
    
    ret["growth_rate"] = ret["growth_rate"].apply(lambda x: x.replace("%",""))
    
    ret["growth_rate"] = pd.to_numeric(ret["growth_rate"],errors="coerece")
    
    ret["year"] = "2015-16"
    return ret

state_df()
```

```python
state_df().to_csv("../clean/for_db/state_sbac_growth_ell.csv",index=False)
```
