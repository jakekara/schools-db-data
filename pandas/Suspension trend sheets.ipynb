```python
import pandas as pd, re
import getschools as gs
```

```python
def get_df(group):
    ret = pd.read_csv("../scripts/noequals/SuspensionRateExport-Trend-" + group + ".csv",header=None)
    
    ret = gs.merge_01(ret)
    
    ret = gs.clean_colnames(ret)
    ret["district_code"] = ret["district_code"].str.replace("\"","")
    ret["school_code"] = ret["school_code"].str.replace("\"","")

    ret["district_code"] = ret["district_code"].astype(str).str.zfill(7)
    ret["school_code"] = ret["school_code"].astype(str).str.zfill(7)
    
    return ret

```

```python
'''SuspensionRateExport-Trend-sped.csv
SuspensionRateExport-Trend-frpl.csv
SuspensionRateExport-Trend-race.csv
SuspensionRateExport-Trend-gender.csv
SuspensionRateExport-Trend-ell.csv
SuspensionRateExport-Trend-all.csv'''
```

```python
groups = "sped","frpl","race","gender","ell","all"

for g in groups:
    get_df(g).to_csv("../clean/suspension-trend-" + g + ".csv", index=False)
```
