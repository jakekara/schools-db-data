```python
def get_df(suff="13-14"):
    year = "20" + suff
        
    srcdir="../scripts/noequals/"
    dstdir="../clean/csv/"
    
    fname = "CCRCourseTaking-"+suff+".csv"
    
    ret = pd.read_csv(srcdir + fname, skiprows=7)

    ret.columns="district","district_code","11th_count","11_pct","12th_count","12th_pct"
    ret["district_code"] = ret["district_code"].str.replace("\"","")
    ret["school_code"] = ret["district_code"] 
    ret.to_csv(dstdir + fname,index=False)
    
    return ret

get_df()
```
