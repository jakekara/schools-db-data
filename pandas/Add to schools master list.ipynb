```python
import pandas as pd
import getschools as gs
```

```python
def get_schools_df():
    ret =  pd.read_csv("../clean/csv/schools.csv")
    
    ret["school_code"] = ret["school_code"].astype(str).str.zfill(7)
    
    return ret

schools = get_schools_df()

get_schools_df().head()

```

```python
def get_school(code):
    code = str(code).zfill(7)
    return schools[schools["school_code"] == code]

get_school(90182)
```

```python
def is_in_schools(code):
    return len(get_school(code)) == 1
```

```python
is_in_schools("90182")
```

```python
is_in_schools("")
```

```python
is_in_schools("0090182")
```

```python
def check_file(fname):
    
    df = pd.read_csv(fname)
    
    return df[df.apply(lambda x: not is_in_schools(x["school_code"]), axis=1)]
    
check_file("../clean/csv/enrollment-trend-ell.csv")
```
