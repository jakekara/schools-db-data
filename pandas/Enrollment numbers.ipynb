```python
year = "2015" # based on the data I have right now
```

```python
def get_df(groupstr,skiprows=4):
    srcdir="../scripts/noequals/"
    dstdir="../clean/csv/"
    fname = "enrollmentSingleYear-"+ groupstr + ".csv"
    ret = pd.read_csv(srcdir + fname,skiprows=skiprows)
    
    print ret.columns
    ret = filldown(ret, "District")
    ret = filldown(ret, "School")
    
    ret = clean_colnames(ret)
    ret["year"] = year
    ret["group"] = groupstr
    
    ret["district_code"] = ret["district_code"].apply(lambda x: re.sub(r'"','',x))
    ret["school_code"] = ret["school_code"].apply(lambda x: re.sub(r'"','',x))

    ret.to_csv(dstdir + fname,index=False)
    return ret
    
get_df("all",skiprows=3)
```
