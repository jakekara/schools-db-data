```python
import pandas as pd
import numpy as np
import getschools as gs
```

```python
def get_df():
    
    ret = pd.read_csv("../clean/csv/suspension-trend-all.csv",dtype={"district_code":str,"school_code":str})
    ret["race"] = "All"
    
    race = pd.read_csv("../clean/csv/suspension-trend-race.csv",dtype={"district_code":str,"school_code":str})
    
    race["race"] = race["race_ethnicity"]
    
    return pd.concat([race, ret])


get_df()
```

```python
def year_df(df,year):
    
    ret = df.copy()

    ret["year"] = year.replace("_","-")
    
    ret["count"] = pd.to_numeric(ret[year + "_count"],errors="coerce")
    ret["pct"] = pd.to_numeric(ret[year + "_pct"],errors="coerce")
    
    ret = gs.add_drg(ret)

    
    return ret
```

```python
df = get_df()

def all_df():
    
    frames = []
    for y in range(11,16):
        year = "20" + str(y) + "_" + str(y + 1)
        frames.append(year_df(df,year))
        
    ret = pd.concat(frames)
    
    
    #print ret.columns
    
    
    ret = ret[[u'district', u'district_code', u'school', u'school_code',
              u'race', u'year',u'count', u'pct',"drg"]]
    
    ret = ret.sort_values(by=["district","school","year","count"],ascending=False)
    
    return ret
        
all_df().to_csv("../clean/for_db/suspensions.csv",index=False)
```
