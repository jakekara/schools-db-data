```python
import pandas as pd
import numpy as np
import getschools as gs
```

```python
def get_df():
    
    ret = pd.read_csv("../scripts/noequals/fte-staffing-levels-w-15-16.csv",
                      skiprows=3,usecols=range(0,9))
    
    ret = gs.clean_colnames(ret)
    #ret = gs.add_exact_matches(ret)
    
    ret["district"] = ret["district"].fillna(method="ffill")
    ret["school"] = ret["school"].fillna(method="ffill")
    
    ret = gs.add_district_matches(ret)
    ret["district_code"] = ret["dist_code"]
    ret = gs.add_exact_matches(ret)
    
    ret = gs.add_drg(ret)

    return ret

get_df()
```

```python
def year_df(df, ycols,colname):

    frames = []
    
    for y in ycols:
        
        new_frame = df.copy()
        
        new_frame["year"] = y.replace("_","-")
        new_frame[colname] = df[y]
        
        frames.append(new_frame)
        
    ret = pd.concat(frames)
    
    cols = filter(lambda x: x not in ycols, ret.columns)
    
    ret = ret[cols]
    
    return ret

#year_df(get_df(), ["2011_12","2012_13","2013_14","2014_15","2015_16"], "count")
```

```python
df = year_df(get_df(), ["2011_12","2012_13","2013_14",
                        "2014_15","2015_16"], "count")
```

```python
total_cache = {}

def total(year, district, school):
    
    if year not in total_cache:
        total_cache[year] = {}
        
    if district not in total_cache[year]:
        total_cache[year][district] = {}
    
    if school in total_cache[year][district]:
        return total_cache[year][district][school]
    
    ret = df.copy()
    
    ret = ret[(ret["district"] == district)\
               & (ret["school"] == school)\
               & (ret["year"] == year)]["count"].sum()
    
    total_cache[year][district][school] = ret
    
    return ret

total("2011-12","Achievement First Hartford Academy Inc. District","Achievement First Hartford Academy Inc.")
    
```

```python
df["total"] = df.apply(lambda x: total(x["year"],
                                       x["district"],
                                       x["school"]),
                       axis=1)
df
```

```python
def drg_report():
    ret = df.copy()
    

    ret = ret.groupby(["drg","year","assignment_category","educator_type"]).agg({
            "pct":np.nanmedian
        })
    
    return ret
    
    
drg_report()
```
