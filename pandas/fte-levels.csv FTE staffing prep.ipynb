```python
from getschools import lookup_exact, lookup_contains, lookup_both, add_exact_matches
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../raw/FTEStaffing-precleaned-with-OO.csv")
```

```python
# cleanup 

df["District"] = df["District"].fillna(method="ffill")
df["School"] = df["School"].fillna(method="ffill")
df.columns = [u'district', u'school', u'category', u'type',
       u'2010-11', u'2011-12', u'2012-13', u'2013-14', u'2014-15',
       u'Unnamed: 9']

df = df[[u'district', u'school', u'category', u'type',
       u'2010-11', u'2011-12', u'2012-13', u'2013-14', u'2014-15']]

df.head()
```

```python
with_codes = add_exact_matches(df)
```
