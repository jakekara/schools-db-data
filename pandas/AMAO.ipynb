```python
import pandas as pd
import numpy as np
import getschools as gs
```

```python
def get_df(year="2013-14"):
    #
    # prog target is 74 percent
    # prof target is 24 percent
    ret = pd.read_excel("../raw/AMAO 2009-14.xlsx",sheetname="SY " + year, skiprows=1)
    
    ret.columns = ["district","progress","proficient","academic_target_met","all_three_met"]
    
    def clean_dist_name(name):
        name = str(name).strip()
        
        if "School District" in name:
            return name
        return name + " School District"
        
    
    ret["district"] = ret["district"].apply(clean_dist_name)
    
    ret = gs.add_district_matches(ret)
    
    ret["district_code"] = ret["dist_code"]
    ret["school_code"] = ret["district_code"]
    ret["year"] = year
    ret = gs.add_drg(ret)
    
    return ret

get_df().head()
```

```python
def get_all_years():
    
    frames = []
    for y in range(10,14):
        year = "20" + str(y) + "-" + str(y+1)
        frames.append(get_df(year=year))
        
    return pd.concat(frames)

get_all_years()
df = get_all_years()  
```
