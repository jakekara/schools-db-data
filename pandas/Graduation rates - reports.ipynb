```python
import pandas as pd
import numpy as np
import getschools as gs
```

```python
def get_df(group):
    
    ret = pd.read_csv("../clean/csv/graduation-" + group + ".csv",dtype={"school_code":str})
    ret = gs.add_district_code(ret)
    ret = gs.add_drg(ret)
    return ret

get_df("all")
```

```python
def year_sheet(df, year, colname="grad_rate",cols=["school_code","drg","district","school"]):
    
    cols += [year]
    ret = df.copy()[cols]
    
    cols = []
    for c in ret.columns:
        #print c
        cols.append(c)

    cols[-1] = colname
    
    #print cols
    
    ret.columns = cols
    ret["year"] = year.replace("_","-")
    ret[colname] = pd.to_numeric(ret[colname],errors="coerce")
    #print ret.columns
    return ret

year_sheet(get_df("all"),"2015_16").head()
```

```python

def trend_frame(group, groups=[]):
    grad_df = get_df(group)
    
    frames = []

    for y in range(11,16):


        year = "20" + str(y) + "_" + str(y + 1)
        #print year
        new_df = year_sheet(grad_df,year,colname="grad_rate",cols=["school_code","district","drg","school"] + groups)
        #print new_df.head()
        frames.append(new_df)
#         frames.append(year_sheet(grad_df,year,colname="grad_rate",cols=["school_code","district","school"]))
        
    #print len(frames)
    return pd.concat(frames)

trend_frame("all").to_csv("../clean/for_db/grad_rates.csv",index=False)
```

```python
def drg_report(group, groups=[]):
    trend_df = trend_frame(group,groups)

    def do_drg_report():
        ret = trend_df

        ret = ret.groupby(["drg","year"] + groups).agg({"grad_rate":np.nanmedian}).reset_index()

        return ret
    
    return do_drg_report()

drg_report("all")#.to_csv("../clean/for_db/drg_grad_rates.csv",index=False)
```

```python
def state_report():
    
    df = pd.read_csv("../raw/state_grad_rates.csv",skiprows=5)
    df = gs.clean_colnames(df)
    
    rows = []
    for y in range(11,16):
        year = "20" + str(y) + "_" + str(y + 1)
        rows.append({
                "grad_rate": df[year][0],
                "year":year.replace("_","-")
            })
    
    return pd.DataFrame(rows)

state_report().to_csv("../clean/for_db/state_grad_rate.csv",index=False)
```
