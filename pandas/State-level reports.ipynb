# Enrollment trend reports

```python
def get_df():
    ret = pd.read_csv("../clean/enrollment-trend-race.csv")
    ret = gs.add_drg(ret)
    
    return ret

```

```python
def summary(df,group_keys=[]):
    
    agg = {}
    
    ret = df.copy()
    
    for c in ret.columns:
        if c in group_keys:
            continue
        ret[c] = pd.to_numeric(ret[c], errors="coerce")
        agg[c] = np.sum
        
    print ret.columns
    
    return ret.groupby(group_keys).agg(agg).reset_index()
    
```

```python
def drg_summary(df):
    return summary(df, ["drg","year"])
```

```python
def state_summary(df):
    return summary(df, ["year"])
```

```python
def enrollment_trend_report(group):
    df = pd.read_csv("../clean/csv/enrollment-trend-" + group + ".csv")
    df["district_code"] = df["district_code"].astype(str).str.zfill(7)
    df["school_code"] = df["school_code"].astype(str).str.zfill(7)
        
    df = gs.add_drg(df) 
    
    state_summary(df).to_csv("../clean/drg_state_reports/state_enrollment_trend_" + group + ".csv",index=False)
    drg_summary(df).to_csv("../clean/drg_state_reports/drg_enrollment_trend_" + group + ".csv",index=False)
   
    drg_summary(df).to_csv("../clean/for_db/drg_enrollment_trend_" + group + ".csv",index=False)

    
    return df

enrollment_trend_report("race")
```

```python
for group in ["ell","gender","grade","lunch","race","sped"]:
    enrollment_trend_report(group)
```

# EdSight state-level summary

EdSight's sum varies slightly from ours -- because it specifically looks at Public Schools.

We do not have school type information for every school, so I'll defer to EdSight for the state numbers.

```python
def clean_race_cols(df):
    race_aliases = {
        "black":["black_not_of_hispanic_origin","black_or_african_american"],
        "white":["white_not_of_hispanic_origin"],
        "asian":[],
        "hispanic":["hispanic_latino","hispanic_latino_of_any_race"],
        "other":["american_indian","american_indian_or_alaska_native","native_hawaiian_or_other_pacific_islander",
                 "not_reported","two_or_more_races"]
    }
    
    ret = df.copy()
    ret["tot_check"] = pd.to_numeric(ret["total"],errors="coerce").fillna(0)
    
    for k in race_aliases:
        print k
        if k not in ret.columns:
            ret[k] = 0
        else:
            ret[k] = pd.to_numeric(ret[k], errors="coerce").fillna(0)
            print "starting ", k, "with", k
        for alias in race_aliases[k]:
            try:
                print "Adding ", alias, " to ", k
                ret[alias] = pd.to_numeric(ret[alias],errors="coerce").fillna(0)
                ret[k] += ret[alias]
            except:
                pass
        ret["tot_check"] -= pd.to_numeric(ret[k],errors="coerce").fillna(0)
        print "tot_check sum", ret["tot_check"].sum()

    
    ret["oth_total"] = pd.to_numeric(ret["total"],errors="coerce").fillna(0)\
    - ret["black"] - ret["white"] - ret["hispanic"] - ret["asian"]


    
    return ret
    
```

```python
def edsight_race_trend():
    df = pd.read_csv("../raw/EnrollmentTrend-state.csv",skiprows=5).transpose()
    df = df.reset_index()
    df.columns = df.iloc[0]
    df = df.drop(0)
    df = gs.clean_colnames(df)
    df.columns = [u'year', u'american_indian_or_alaska_native', u'asian',
       u'black_or_african_american', u'hispanic_latino_of_any_race',
       u'native_hawaiian_or_other_pacific_islander', u'not_reported',
       u'two_or_more_races', u'white', u'total']
    
    
    return clean_race_cols(df)

edsight_race_trend().columns
```
