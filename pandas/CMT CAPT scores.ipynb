```python
import pandas as pd
import getschools as gs
```

```python
year = "2015-16"

def get_df(suff="all"):
    fname="CMTCAPTScience-"+suff+".csv"
    srcdir="../scripts/noequals/"
    dstdir="../clean/csv/"
    
    
    ret = pd.read_csv(srcdir + fname,skiprows=3,header=None)
        
    row0 = ret.iloc[0].fillna(method="ffill")
    row1 = ret.iloc[1].fillna("")
    
    cols = []
    for i in range(len(row0)):
        if row0[i] is None:
            cols.append(str(row1[i]))
        else:
            cols.append(str(row0[i]) + "_" + str(row1[i]))

    cols = map(lambda x: x.replace("nan_",""), cols)

    ret = ret.drop(0)
    ret = ret.drop(1)
    ret.columns = cols
    
    ret = gs.clean_colnames(ret)
    
    cols = []
    
    for c in ret.columns:
        cols.append(str(c))
    
    cols[-1] = "average_ss"
    cols[3] = "school_code"
    
    ret.columns = cols
    ret["year"] = year
    
    ret["district_code"] = ret["district_code"].str.replace("\"","")
    ret["school_code"] = ret["school_code"].str.replace("\"","")

    ret.to_csv(dstdir + fname,index=False)
    
    return ret

    
get_df()
```
